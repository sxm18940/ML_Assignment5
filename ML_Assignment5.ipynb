```python
X = dataframe_pd.drop('class',axis=1).values
Y = dataframe_pd['class'].values
```

```python
# a. Perform Scaling
#Scaling Data
scaler = StandardScaler()
X_Scale = scaler.fit_transform(X)
```

```python
# b. Apply PCA (k=3)
# Apply PCA with k =3
pca3 = PCA(n_components=3)
principalComponents = pca3.fit_transform(X_Scale)
principalDataframe = pd.DataFrame(data = principalComponents, columns = ['principal_component1', 'principal_component2','Principal_Component3'])
finalDataframe = pd.concat([principalDataframe, dataframe_pd[['class']]], axis = 1)
finalDataframe.head()
```

```python
X = finalDataframe.drop('class',axis=1).values
Y = finalDataframe['class'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=0)
```

```python
# c. Use SVM to report performance
from sklearn.svm import SVC
SVMClassifier = SVC()
SVMClassifier.fit(X_train, Y_train)
y_pred = SVMClassifier.predict(X_test)
# predictions by the classifier
print(classification_report(Y_test, y_pred, zero_division=1))
print(confusion_matrix(Y_test, y_pred))
# Accuracy score
glass_acc_svc = accuracy_score(y_pred,Y_test)
print('Accuracy is',glass_acc_svc)
#silhouette Score Calculation
score = metrics.silhouette_score(X_test, y_pred)
print("Silhouette Score is: ",score) 
```

#Question-4: Briefly identify the difference between PCA and LDA 
Principal Component Analysis (PCA) and Linear Discriminant Analysis (LDA) are two of the most popular dimensionality reduction techniques. 
Both methods are used to reduce the number of features in the dataset while retaining as much information as possible.
PCA: Principal Component Analysis (PCA) works by identifying the direction (component) that maximizes the variance of a data set. 
That is, it tries to find a linear combination of features that captures as much variance as possible. 
The first component captures the maximum variance, the second component is orthogonal to the first and captures the remaining variance. 
PCA is a useful technique for dimensionality reduction when there is a linear relationship between features in the data
LDA: Linear Discriminant Analysis (LDA) is another linear transformation technique used for dimensionality reduction. 
However, unlike PCA, LDA is a method of supervised learning. That is, class labels are taken into account when finding the direction of maximum variance. 
This makes LDA particularly suitable for classification tasks where we want to maximize class separability.
PCA is an unsupervised learning algorithm and LDA is a supervised learning algorithm. 
This means that PCA finds the direction of maximum variance, independent of class labels, and LDA finds the direction of maximum class separability.
In general, for classification purposes, LDA should be used.
If your goal is just to find patterns in the data (not classify) then PCA might work better.
If noise dominates the data set, LDA may give better results even if the goal is not classification. 
Because LDA makes stronger assumptions about the structure of the data, it often performs better than PCA when the dataset fits those assumptions, but performs worse when it doesn't.
